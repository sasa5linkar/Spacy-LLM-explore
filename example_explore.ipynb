```python
import spacy
from spacy.tokens import DocBin

# Load the DocBin from disk
docbin = DocBin().from_disk("Corpus/latSRP18750_JakovI_VasaRespekt_C01.spacy")

# Load the spaCy model
nlp = spacy.blank("sr")

# Iterate over each document in the DocBin
for doc in docbin.get_docs(nlp.vocab):
    # Print the evething of each document
    print(doc.text)
    print("Tokens:", [token.text for token in doc])
    print("Named Entities:", [(ent.text, ent.label_) for ent in doc.ents])
    print("Lemma:", [token.lemma_ for token in doc])
    print("POS:", [token.pos_ for token in doc])
    print("Tagger:", [token.tag_ for token in doc])
    print("Dependency:", [token.dep_ for token in doc])
    print("Shape:", [token.shape_ for token in doc])
    print("Alpha:", [token.is_alpha for token in doc])
    print("Stop:", [token.is_stop for token in doc])
    print("Is OOV:", [token.is_oov for token in doc])
    print("Is Space:", [token.is_space for token in doc])
    print("Is Punct:", [token.is_punct for token in doc])
    print("Is Digit:", [token.is_digit for token in doc])
    print("Is Lower:", [token.is_lower for token in doc])
    print("Is Upper:", [token.is_upper for token in doc])
    print("Is Title:", [token.is_title for token in doc])
    print("Is Left Punct:", [token.is_left_punct for token in doc])
    print("Is Right Punct:", [token.is_right_punct for token in doc])
    print("Is sent start:", [token.is_sent_start for token in doc])
    print("Is sent end:", [token.is_sent_end for token in doc])
    print("Morph:", [token.morph for token in doc])
    print("Norm:", [token.norm_ for token in doc])
    print("Prefix:", [token.prefix_ for token in doc])
    print("Suffix:", [token.suffix_ for token in doc])
    print("Ent IOB:", [token.ent_iob_ for token in doc])
    print("Ent Type:", [token.ent_type_ for token in doc])



    print()
```
